```python
import pandas as pd 
url_1 = ' https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url_2 = ' https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url_3 = ' https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

data_1 = pd.read_csv(url_1)
data_2 = pd.read_csv(url_2)
data_3 = pd.read_csv(url_3)

data_1
```

```python
col_name_1 = ['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount']
col_name_2 = ['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount', 'policy_type', 'vehicle_class']
col_name_3 = ['customer', 'state', 'customer_lifetime_value', 'education', 'gender', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'total_claim_amount', 'vehicle_class']


data_1.columns = col_name_1
data_2.columns = col_name_2
data_3.columns = col_name_3
```

```python
def gender_fix(label):
     if str(label)in ['M', 'Male'] :  
        return 'M'
     elif str(label)in ['F','Femal']:
        return 'F'
     else:
        return label
    
def state_fix(state):
    if str(state) == 'AL' :  
        return 'Alabama'
    elif str(state) == 'AK':
        return 'Alaska'
    elif str(state) == 'AZ':
        return 'Arizona'
    elif str(state) == 'AR':
        return 'Arkansas'
    elif str(state) == 'AS':
        return 'American Samoa'
    elif str(state) == 'CA':
        return 'California'
    elif str(state) == 'Cali':
        return 'California'
    elif str(state) == 'CO':
        return 'Colorado'
    elif str(state) == 'CT':
        return 'Connecticut'
    elif str(state) == 'DE':
        return 'Delaware'
    elif str(state) == 'DC':
        return 'District of Columbia'
    elif str(state) == 'FL':
        return 'Florida'
    elif str(state) == 'GA':
        return 'Georgia'
    elif str(state) == 'GU':
        return 'Guam'
    elif str(state) == 'HI':
        return 'Hawaii'
    elif str(state) == 'ID':
        return 'Idaho'
    elif str(state) == 'IL':
        return 'Illinois'
    elif str(state) == 'IN':
        return 'Indiana'
    elif str(state) == 'IA':
        return 'Iowa'
    elif str(state) == 'KS':
        return 'Kansas'
    elif str(state) == 'KY':
        return 'Kentucky'
    elif str(state) == 'LA':
        return 'Louisiana'
    elif str(state) == 'ME':
        return 'Mane'
    elif str(state) == 'MD':
        return 'Maryland'
    elif str(state) == 'MA':
        return 'Massachusetts'
    elif str(state) == 'MI':
        return 'Michigan'
    elif str(state) == 'MN':
        return 'Minnesota'
    elif str(state) == 'MS':
        return 'Mississippi'
    elif str(state) == 'MO':
        return 'Missouri'
    elif str(state) == 'MT':
        return 'Montana'
    elif str(state) == 'NE':
        return 'Nebraska'
    elif str(state) == 'NV':
        return 'Nevada'
    elif str(state) == 'NH':
        return 'New Hampshire'
    elif str(state) == 'NJ':
        return 'New Jersey'
    elif str(state) == 'NM':
        return 'New Mexcio'
    elif str(state) == 'NY':
        return 'New York'
    elif str(state) == 'NC':
        return 'North Carolina'
    elif str(state) == 'ND':
        return 'North Dokata'
    elif str(state) == 'MP':
        return 'Northern Mariana Islands'
    elif str(state) == 'OH':
        return 'Ohio'
    elif str(state) == 'OK':
        return 'Oklahoma'
    elif str(state) == 'OR':
        return 'Oregon'
    elif str(state) == 'PA':
        return 'Pennsylvania'
    elif str(state) == 'PR':
        return 'Puerto Rico'
    elif str(state) == 'RI':
        return 'Rhode Island'
    elif str(state) == 'SC':
        return 'South Carolina'
    elif str(state) == 'SD':
        return 'South Dakota'
    elif str(state) == 'TN':
        return 'Tennessee'
    elif str(state) == 'TX':
        return 'Texas'
    elif str(state) == 'TT':
        return 'Trust Territories'
    elif str(state) == 'UT':
        return 'Utah'
    elif str(state) == 'VT':
        return 'Vermont'
    elif str(state) == 'VA':
        return 'Virginia'
    elif str(state) == 'VI':
        return 'Virgin Islands'
    elif str(state) == 'WA':
        return 'Washington'
    elif str(state) == 'WV':
        return 'West Virginia'
    elif str(state) == 'WI':
        return 'Wisconsin'
    elif str(state) == 'WY':
        return 'Wyoming'
        
    return state

def ed_fix(level):
     if str(level) == 'Bachelors' :  
        return 'Bachelor'
     elif str(level) == 'Masters':
        return 'Master'
     else:
        return level

def car_fix(clas):
     if str(clas) in ['Sports Car', "Luxury SUV", "Luxury Car"] :  
        return "Luxury"
     else:
        return clas

def clv_fix(val):
    if val:
        return str(val).rstrip('%')
    else:
        return val
```

```python
data_1['gender'] = data_1['gender'].apply(gender_fix)    
data_1['state'] = data_1['state'].apply(state_fix)    
data_1['education'] = data_1['education'].apply(ed_fix)    
data_1['vehicle_class'] = data_1['vehicle_class'].apply(car_fix)
data_1['customer_lifetime_value'] = data_1['customer_lifetime_value'].apply(clv_fix)

data_2['gender'] = data_2['gender'].apply(gender_fix)    
data_2['state'] = data_2['state'].apply(state_fix)    
data_2['education'] = data_2['education'].apply(ed_fix)    
data_2['vehicle_class'] = data_2['vehicle_class'].apply(car_fix)
data_2['customer_lifetime_value'] = data_2['customer_lifetime_value'].apply(clv_fix)

data_3['gender'] = data_3['gender'].apply(gender_fix)    
data_3['state'] = data_3['state'].apply(state_fix)    
data_3['education'] = data_3['education'].apply(ed_fix)    
data_3['vehicle_class'] = data_3['vehicle_class'].apply(car_fix)
data_3['customer_lifetime_value'] = data_3['customer_lifetime_value'].apply(clv_fix)
```
